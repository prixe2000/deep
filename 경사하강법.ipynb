```python
from sklearn.datasets import load_diabetes
diabetes=load_diabetes()
import matplotlib.pyplot as plt
x=diabetes.data[:,2]
y=diabetes.target
```

```python
w=1
b=1
y_hat=x[0]*w+b
```

```python
for i in range(1, 100): #주어진 훈련데이터로 한 단위 작업을 진행하는 것을 에보크라고 부름 현재 100번의 에보크를 반복함
  for x_i, y_i in zip(x,y): #zip은 배열의 각 값을 꺼내주는 역할을 함
    y_hat=x_i*w+b
    err=y_i-y_hat
    w_rate=x_i
    w=w+w_rate*err
    b=b+1*err
    print(w,b)
```

```python
plt.scatter(x,y)
pt1=(-0.1, -0.1*w+b)
pt2=(0.15,0.15*w+b)
plt.plot([pt1[0],pt2[0]],[pt1[1],pt2[1]])
plt.show()
```

```python
x_new=0.18
y_pred=x_new*w+b
plt.scatter(x,y)
plt.scatter(x_new,y_pred)
plt.show()
print(y_pred)
```

<오차 역전파를 통한 선형 회기분석>

w와 b의 임의의 값 1,1로 초기화하고 훈련데이터 샘플을 하나씩 대입하여 y와 y^의 오차를 구한다.
1에서 구한 오차를 w와 b의 변화에 따른 y변화율에 곱해 w와 b를 업데이트한다.
y^이 y보다 커지면 오차는 음수가 되어 자동으로 w와b가 줄어드는 방향으로 업데이트됨
반대로 y^이y보다 작으면 오차는 야수가 되고 w와 b는 더 커지도록 업데이트 됨
